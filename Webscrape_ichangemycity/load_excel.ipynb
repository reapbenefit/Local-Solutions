### Import libraries

```
import pandas as pd
import requests
from bs4 import BeautifulSoup
from zipfile import ZipFile 
from io import BytesIO
```

### View html

```
url = 'https://www.ichangemycity.com/mcmb2020'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

### Download xlsx of all wards

```
files_list=[]

for i in range(1,199):
  href = "https://www.ichangemycity.com/theme/files/yellowspots/Ward_"+str(i)+".xlsx"
  files_list.append(href)

output = pd.DataFrame()
for file_name in files_list[:]:
    temp_df = pd.read_excel(file_name)
    output = output.append(temp_df)
    print(file_name)

output.to_excel("ichangemycity.xlsx")
```
